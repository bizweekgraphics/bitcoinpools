```python
# Render our plots inline
%matplotlib inline

import pandas as pd
import matplotlib.pyplot as plt

pd.set_option('display.mpl_style', 'default') # Make the graphs a bit prettier
plt.rcParams['figure.figsize'] = (15, 5)
```

```python
rawBlocksWeekly = pd.read_csv('weeklyBlockSolves.csv', parse_dates=['Date'])
rawBlocksDaily = pd.read_csv('dailyBlockSolves.csv', parse_dates=['Date'])

# Placeholder for interpolating weekly and daily ("adaptive resampling")
blocksFrame = rawBlocksWeekly
```

```python
import urllib2

# Fetching historical prices from Coindesk. API docs at http://www.coindesk.com/api/
url = 'http://api.coindesk.com/v1/bpi/historical/close.json?start=2011-02-27&end=2014-06-25'
response = urllib2.urlopen(url)
pricesJSON = response.read()
prices = pd.read_json(pricesJSON)
```

```python
# Nest raw blocks 
pools = blocksFrame.groupby('Pool').groups.keys()
dates = blocksFrame.groupby('Date').groups.keys()

blocksByPool = []

for pool in pools:
    poolBlocks = {'name': pool}
    poolBlocks['values'] = []
    for date in dates:
        # datum = blocksFrame[blocksFrame['Pool'] == pool]
        datum = {'date': date, 'y': 0, 'blocks': 0, 'revenue': 0}
        poolBlocks['values'].append(datum)
    blocksByPool.append(poolBlocks)

blocksByPool
```
